```python
import cv2
import face_recognition
import numpy as np
import time

# Load the reference image and encode the face
reference_image_path = 'reference_image.jpg'
reference_image = face_recognition.load_image_file(reference_image_path)
reference_encoding = face_recognition.face_encodings(reference_image)[0]

# Initialize variables
process_this_frame = True

# Start video capture from the webcam
video_capture = cv2.VideoCapture(0)

if not video_capture.isOpened():
    print("Error: Could not open webcam.")
    exit()

while True:
    # Grab a single frame of video
    ret, frame = video_capture.read()

    # Check if frame is captured correctly
    if not ret or frame is None:
        print("Error: Frame not captured correctly.")
        break

    # Resize frame of video to 1/4 size for faster face recognition processing
    small_frame = cv2.resize(frame, (0, 0), fx=0.25, fy=0.25)

    # Convert the image from BGR color (which OpenCV uses) to RGB color (which face_recognition uses)
    rgb_small_frame = np.ascontiguousarray(small_frame[:, :, ::-1])

    # Only process every other frame of video to save time
    if process_this_frame:
        # Find all the faces and face encodings in the current frame of video
        face_locations = face_recognition.face_locations(rgb_small_frame)
        face_encodings = face_recognition.face_encodings(rgb_small_frame, face_locations)

        for face_encoding in face_encodings:
            # Check if the face matches the reference face
            matches = face_recognition.compare_faces([reference_encoding], face_encoding)
            face_distances = face_recognition.face_distance([reference_encoding], face_encoding)
            best_match_index = np.argmin(face_distances)
            if matches[best_match_index]:
                print("Face recognized!")
                # Capture and save the frame when the person is recognized
                timestamp = time.strftime("%Y%m%d-%H%M%S")
                cv2.imwrite(f'captured_{timestamp}.jpg', frame)
                print(f"Image saved as captured_{timestamp}.jpg")

    process_this_frame = not process_this_frame

    # Display the resulting frame
    cv2.imshow('Video', frame)

    # Hit 'q' on the keyboard to quit!
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release handle to the webcam
video_capture.release()
cv2.destroyAllWindows()
```
