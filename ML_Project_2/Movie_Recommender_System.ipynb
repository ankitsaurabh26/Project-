```python
def convert_cast(obj):
    Output_List = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter!=3:
            Output_List.append(i['name'])
            counter+=1
        else:
            break 
    return Output_List
```

```python
movies.cast = movies.cast.apply(convert_cast)
```

```python
# I want "job": "Director"

def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies.crew.apply(fetch_director)
```

```python
movies['director'] = movies.crew.apply(fetch_director)
```

**Stemming**

Stemming is the process of reducing a word to its base form, or stem, which can then be used with prefixes, suffixes, or the roots of words, known as "lemmas". For example, "history" and "historical" can be stemmed to "histori". Stemming can be used to convert ['loved','loves','love'] => ['love','love','love']

```python
!pip install nltk
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
ps.stem('loving')
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title']== movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
